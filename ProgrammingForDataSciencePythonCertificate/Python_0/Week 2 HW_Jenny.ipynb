Question 1
PART 1: (2 points)

a. What is the average US Gross of movies in the dataset?

b. How many movies in the dataset have budgets greater than $20 million?

PART 2: (3 points)

c. How many movies were released by each film distributor? (Hint: this could be a good place to use dictionaries)

```python
f = open("MovieData.csv")
rows = []
usGross = []
def toInt(str):
    try:
        return int(str)
    except:
        return None
def getSum(myList):
    result = 0
    for a in myList:
        result += a
    return result
def getMean(myList):
    return getSum(myList)/len(myList)
f.seek(0) # go to the start of data file
f.readline() # skip the title row
for line in f:
    temp = line.split('\t')
    rows.append(temp)
    val = toInt(temp[4])
    if val is not None: # ignore missing value row
        usGross.append(val)
f.close()
print("The average US Gross of movies in the dataset = ", getMean(usGross))
```

```python
result = 0
for row in rows:
    if toInt(row[3]) > 20000000:
        result += 1
print(result,"movies in the dataset have budgets greater than $20 million")
```

```python
movieCount = {}
for row in rows:
    if row[2] in movieCount:
        movieCount[row[2]] += 1
    else:
        movieCount[row[2]] = 1
print("Distributor : #Movies")
for key in sorted(movieCount.keys()): # sort by film distributor name
    print(key, ":", movieCount[key]) 
```

Question 2
PART 1: (2 points)

Load the earthquake data in QuakeData.csv, and use it to answer the following questions:

a. How many earthquakes are in the dataset?

PART 2: (3 points)

b. What is the average magnitude of earthquakes in the dataset?

c. The DateTime format in this dataset is a bit trickier than it was for movies. Try to parse it into a datetime object. Do most earthquakes happen between midnight and noon, or noon to midnight?

Hint
The date string format here is more complicated than in the previous problem. Here's an example date string:

2012-01-01T00:30:08.770+00
Notice that the first four characters are the year; that means we can extract it by subsetting. If the date string is stored in a string variable date_str, getting the first four characters and converting them can be done like this:

year = int(date_str[0:4])
The same process can extract the month, day, and the hours, minutes, and seconds -- just adjust the subset to the correct part of the string


```python
def convert(record):
    try:
        return float(record)
    except:
        return record
f = open('QuakeData.csv')
rows = []
f.readline()
for line in f:
    line = line.split(",") # Split on commas
    line = [convert(x) for x in line] # Convert each entry
    rows.append(line)
f.close()
print(len(rows), "earthquakes are in the dataset")
```

```python
magnitude = [row[4] for row in rows] # list of magnitude
print("The average magnitude of earthquakes = ", getMean(magnitude))
```

```python
import datetime as dt

def makeDate(str):
    y = int(str[0:4])
    m = int(str[5:7])
    d = int(str[8:10])
    h = int(str[11:13])
    min = int(str[14:16])
    s = int(str[17:19])
    return dt.datetime(y, m, d, h, min, s)
datetimes = [makeDate(row[0]) for row in rows] # list of datetime object of first column time str
hours = [date.hour for date in datetimes]
midnightToNoon = len([x for x in hours if x >= 0 and x < 12])
noonToMidnight = len([x for x in hours if x >= 12 and x <= 23])
print(midnightToNoon, "earthquakes happen between midnight and noon")
print(noonToMidnight, "earthquakes happen between noon and midnight")
if midnightToNoon > noonToMidnight:
    print("Earthquakes happen between midnight and noon is more than between noon and midnight")
else:
    print("Earthquakes happen between noon and midnight is more than between midnight and noon")
```
