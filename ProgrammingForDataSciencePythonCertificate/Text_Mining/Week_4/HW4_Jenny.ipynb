# (7 points): hashtag predictor. parse, tokens, dictionary and vector to train a classifier for each hashtag. naïve bayes classifier. ‘create_training_data’ and ‘create_testing_data’, ‘compute_accuracy’

```python
(train_data,train_labels,vocab) = create_training_data('TextMining_twitter_train_wk_4.txt')
(test_data, test_labels) = create_testing_data('TextMining_twitter_test_wk_4.txt', vocab)
```

```python
train_df = pd.concat([pd.DataFrame(train_data), pd.DataFrame(train_labels, columns=['label'])], axis = 1)
train_df.head()
```

```python
test_df = pd.concat([pd.DataFrame(test_data), pd.DataFrame(test_labels, columns=['label'])], axis = 1)
test_df.head()
```

```python
nbClassifier = train_naive_bayes(train_df.drop(columns=['label']), train_df['label'])
nbPred = evaluate_classifier(nbClassifier, test_df.drop(columns=['label']))
compute_accuracy(nbPred, test_df['label'])
```
